## Introducción

Este proyecto consiste en construir un modelo de Machine Learning capaz de predecir el precio de anuncios de Airbnb a partir de distintas características del alojamiento y del anfitrión.

El objetivo es estimar automáticamente el precio que podría tener una propiedad, para ello se ha utilizado un dataset real de Airbnb. 

El proceso incluye: carga y limpieza de datos, análisis exploratorio, selección de variables mediante Random Forest, generación de nuevas variables, validación cruzada del modelo y predicción final sobre un conjunto de test.

El modelo final ha logrado un error medio (RMSE) de aproximadamente 9 €.

```python
# Librerias
import pandas as pd
import numpy  as np  

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

## 1. Preparación de datos: División train/test

```python
# Cargar dataset
full_df = pd.read_csv("airbnb-listings-extract.csv", sep=None, engine="python")

# Dividir en train/test
train_df, test_df = train_test_split(full_df, test_size=0.2, shuffle=True, random_state=42)

print(f'Dimensiones del dataset de training: {train_df.shape}')
print(f'Dimensiones del dataset de test: {test_df.shape}')

# Guardar en csv separados
train_df.to_csv('./airbnb_train.csv', sep=';', decimal='.', index=False)
test_df.to_csv('./airbnb_test.csv', sep=';', decimal='.', index=False)

airbnb_data = pd.read_csv('./airbnb_train.csv', sep=';', decimal='.')
airbnb_test = pd.read_csv('./airbnb_test.csv', sep=';', decimal='.')
airbnb_data.head(5).T
```

## 3. Preprocesamiento

### 3a. Eliminación de variables

Antes de este punto se ha realizado ya la eliminación de:

- **Variables sin valor predictivo** (como IDs, URLs o imágenes), eliminadas manualmente. -> *columns_to_drop_1*
- **Columnas con más del 70% de valores nulos**, identificadas en el análisis exploratorio. -> *columns_to_drop_2*


Se procederá a la eliminación de:
* **Variables con menor correlación <=0,05 con el precio:**

  - Maximum Nights
  - Review Scores Checkin
  - Longitude 
  - Latitude
  - Number of Reviews
  - Host Response Rate
  - Review Scores Communication
  - Review Scores Value

```python
low_corr_vars = ["Maximum Nights", "Review Scores Checkin", "Longitude", "Latitude", "Number of Reviews", "Host Response Rate", "Review Scores Communication", "Review Scores Value"]

airbnb_data = airbnb_data.drop(columns=low_corr_vars) # Train

airbnb_test = airbnb_test.drop(columns=low_corr_vars) # Test
```

* **Variables redundantes:**

    Muy alta correlación entre las siguientes variables, lo que indica redundancia de información. 

    - Host Listings Count
    - Host Total Listings Count

```python
redundant_vars = ["Host Listings Count", "Host Total Listings Count"]

airbnb_data = airbnb_data.drop(columns=redundant_vars) # Train

airbnb_test = airbnb_test.drop(columns=redundant_vars) # Test
```

* **Otras variables sin valor predictivo:**
  
    Contienen descripciones de texto libre, datos redundantes o información incompleta. La decisión se basa en la revisión de correlación, porcentaje de nulos y utilidad práctica para el modelo.


```python
columns_to_drop_3 = [
    'Name',
    'Summary',
    'Description',
    'Experiences Offered',
    'Neighborhood Overview',
    'Notes',
    'Transit',
    'Access',
    'Interaction',
    'House Rules',
    'Host About',
    'Host Name',
    'Host Location',
    'Host Neighbourhood',
    'Street',
    'Neighbourhood',
    'Neighbourhood Cleansed',
    'City',
    'State',
    'Zipcode'
]

airbnb_data.drop(columns=columns_to_drop_3, inplace=True) # Train

airbnb_test.drop(columns=columns_to_drop_3, inplace=True) # Test
```

* **Variables de tipo object**

  Estas variables crearán conflicto al aplicar el Random Forest

```python
# Columnas de tipo object
airbnb_data.select_dtypes(include='object').columns.tolist()
```

```python
columns_object = [
    'Space', 'Host Verifications', 'Market', 'Smart Location',
    'Country Code', 'Country', 'Amenities', 'Calendar Updated',
    'Calendar last Scraped', 'First Review', 'Last Review',
    'Geolocation', 'Features'
]

airbnb_data.drop(columns=columns_object, inplace=True) # Train

airbnb_test.drop(columns=columns_object, inplace=True) # Test
```

```python
# Transformamos la fecha "Host Since" en antigüedad del host en días
#Train
airbnb_data['Host Since'] = pd.to_datetime(airbnb_data['Host Since'], errors='coerce')
airbnb_data['host_antiguedad_dias'] = (pd.to_datetime('2021-01-01') - airbnb_data['Host Since']).dt.days
airbnb_data.drop(columns=['Host Since'], inplace=True)

#Test
airbnb_test['Host Since'] = pd.to_datetime(airbnb_test['Host Since'], errors='coerce')
airbnb_test['host_antiguedad_dias'] = (pd.to_datetime('2021-01-01') - airbnb_test['Host Since']).dt.days
airbnb_test.drop(columns=['Host Since'], inplace=True)

# Imputación de nulos
cols_mediana = [
    'Bathrooms', 'Bedrooms', 'Beds', 'Review Scores Rating',
    'Review Scores Accuracy', 'Review Scores Cleanliness',
    'Review Scores Location', 'Reviews per Month',
    'Cleaning Fee', 'Security Deposit', 'host_antiguedad_dias',
    'Calculated host listings count'
]

# Train
for col in cols_mediana: 
    if col in airbnb_data.columns:
        airbnb_data[col].fillna(airbnb_data[col].median(), inplace=True)

# Test
for col in cols_mediana:
    if col in airbnb_test.columns:
        airbnb_test[col].fillna(airbnb_test[col].median(), inplace=True)

# Imputación con moda para columnas categóricas
airbnb_data['Host Response Time'].fillna(airbnb_data['Host Response Time'].mode()[0], inplace=True) # Train
airbnb_test['Host Response Time'].fillna(airbnb_test['Host Response Time'].mode()[0], inplace=True) # Test

# Codificación de variables categóricas
categorical_vars = ['Property Type', 'Room Type', 'Bed Type', 'Cancellation Policy',
                    'Neighbourhood Group Cleansed', 'Host Response Time']

airbnb_data = pd.get_dummies(airbnb_data, columns=categorical_vars, drop_first=True) # Train
airbnb_test = pd.get_dummies(airbnb_test, columns=categorical_vars, drop_first=True) # Test
```

**Random Forest**

```python
# Eliminamos posibles nulos en Price
airbnb_data = airbnb_data.dropna(subset=['Price'])

# Separamos X e y
X = airbnb_data.drop('Price', axis=1)
y = airbnb_data['Price']

# Entrenamos el modelo
rf = RandomForestRegressor(n_estimators=100, random_state=0)
rf.fit(X, y)

# Variables con Importancia
importancia_variables = pd.DataFrame({
    'Variable': X.columns,
    'Importancia': rf.feature_importances_
})

# Ordenar de mayor a menor importancia
importancia_variables.sort_values(by='Importancia', ascending=False, inplace=True)

importancia_variables.head(20)
```

### 3b. Generación de variables

```python
# Train
airbnb_data['beds_squared'] = airbnb_data['Beds']**2
airbnb_data['bed_bath_product'] = airbnb_data['Beds'] * airbnb_data['Bathrooms']
airbnb_data['price_per_guest'] = airbnb_data['Price'] / airbnb_data['Accommodates']

# Test
airbnb_test['beds_squared'] = airbnb_test['Beds']**2
airbnb_test['bed_bath_product'] = airbnb_test['Beds'] * airbnb_test['Bathrooms']
airbnb_test['price_per_guest'] = airbnb_test['Price'] / airbnb_test['Accommodates']
```

**Repetición del Random Forest para ver si las nuevas variables las considera importantes**

```python
# Eliminamos posibles nulos en la nueva variable price_per_guest
airbnb_data = airbnb_data.dropna(subset=['price_per_guest'])

# Separamos X e y
X = airbnb_data.drop('Price', axis=1)
y = airbnb_data['Price']

# Entrenamos el modelo
rf = RandomForestRegressor(n_estimators=100, random_state=0)
rf.fit(X, y)

# Variables con Importancia
importancia_variables = pd.DataFrame({
    'Variable': X.columns,
    'Importancia': rf.feature_importances_
})

importancia_variables.sort_values(by='Importancia', ascending=False, inplace=True)

importancia_variables.head(20)
```

Tras generar nuevas variables y repetir el análisis de importancia con Random Forest, observamos que **price_per_guest** se convirtió en la variable más importante con mucha diferencia, y **bed_bath_product** también destacó . Esto indica que las variables aportan valor predictivo y deben ser incluidas en el modelo final.


```python
top_vars = importancia_variables.head(15)['Variable'].tolist()
top_vars
```

### Cross Validation

```python
X = airbnb_data[top_vars]
y = airbnb_data['Price']

# Modelo de Random Forest
modelo = RandomForestRegressor(n_estimators=100, random_state=0)

# Validación cruzada con 5 particiones y error cuadrático medio negativo
scores = cross_val_score(modelo, X, y, cv=5, scoring='neg_mean_squared_error')

# Convertimos a RMSE y mostramos la media
rmse_scores = np.sqrt(-scores)
print(f"RMSE medio: {rmse_scores.mean():.2f}")
```

### Probar con Test

```python
# Definir X_train e y_train
X_train = airbnb_data[top_vars]
y_train = airbnb_data['Price']

# Definir X_test
X_test = airbnb_test[top_vars]

# Entrenar modelo con todo el train
modelo_final = RandomForestRegressor(n_estimators=100, random_state=0)
modelo_final.fit(X_train, y_train)

# Predecir en test
predicciones = modelo_final.predict(X_test)

# Mostramos las 10 primeras predicciones
print(predicciones[:10])
```

```python
plt.scatter(airbnb_test['Price'], predicciones, alpha=0.5)
plt.xlabel("Precio real")
plt.ylabel("Predicción")
plt.title("Predicción vs Real")
plt.plot([0, 500], [0, 500], '--', color='red')  # línea de referencia perfecta
plt.show()
```

## Conclusiones

- El preprocesamiento aplicado al conjunto de test fue idéntico al del conjunto de entrenamiento, asegurando que el modelo recibió datos en el mismo formato.
  
- Se ha entrenado un modelo de regresión utilizando Random Forest.

- La selección de variables se realizó aplicando Random Forest sobre el conjunto de train.

- Se generaron nuevas variables que mejoraron la capacidad predictiva del modelo. La más relevante fue **price_per_guest**.

- El modelo fue evaluado mediante validación cruzada, obteniendo un **RMSE medio de 9.02 euros**, lo que indica que, de media, el modelo puede equivocarse en ±9 € al predecir el precio de un alojamiento.

- Las primeras predicciones se encuentran valores como 70 €, 130 €, y 180.23 €, que corresponden a propiedades del conjunto de test.
